## Quantum Circuits

Quantum Circuits are made by applying Quantum Gates on the qubits that hold either zero or one. The quantum gates consist of X gate, Y gate, Z gate, Hadamard gate, Controlled-X gate, Controlled-Z gate, SWAP gate, Toffoli gate and Phase gate. Each of these gates can be represented in a matrix form. These gates can be arranged and combined in specific sequences creating an algorithm for computing simple to difficult tasks and each qubit renders a value of one or zero in the end when measured.

The course material helped me learn the math behind these gates and how a quantum gate changes the value of a qubit mathematically when applied.

## Future of Quantum Computing

Quantum computers have so many emerging applications and could change the world. It could transform medicine, break encryption and revolutionise communications and artificial intelligence. A machine with fewer than 100 stable qubits should exhibit quantum supremacy, which means it could solve problems in a few seconds that would take any classical machine thousands of years to crunch[3].

How is Quantum Computing so fast ? If you ask a normal computer to figure its way out of a maze, it will try every single branch in turn, ruling them all out individually until it finds the right one. A quantum computer can go down every path of the maze at once. It can hold uncertainty in its head[4]. This idea could change the looping and time complexity concept in the current object-oriented programming approaches. 

Quantum computers will fast track the search of new elements and materials by simulating them and performing reactions with other simulated materials to check for results. Quantum computers could have potentially fast tracked the search of covid vaccines by simulating the virus molecule and running experiments of drug compositions to observe the effects of the those drugs on the virus molecule[5]. Right now, supercomputers can only analyse the most basic molecules. But quantum computers operate using the same quantum properties as the molecules theyre trying to simulate. They should have no problem handling even the most complicated reactions[4]. Scientists hope that quantum computers can find cure for many other diseases whose treatment is not available at the moment.

With fast computation comes some challanges as well. Quantum computers can break encryptions putting a lot of data at risk of getting exposed in public domain and potentially falling in wrong hands. To prevent this from happening, quantum encryption will have to be employed to take care of the data. Another challenge would be for cryptocurrencies, with the use of quantum computer, crypto mining would be much faster and could potentially destabilize the whole crypto blockchain.  

Once quantum machines become more powerful and more widely available, countless other applications will suggest themselves. This will result in a surge in demand of quantum developers and open a new market.
